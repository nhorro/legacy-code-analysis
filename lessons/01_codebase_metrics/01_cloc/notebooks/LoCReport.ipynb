# Métricas con cloc

```python
import pandas as pd
```

```python
df = pd.read_csv("../results/cloc_metrics.csv")
df = df.iloc[:-1,:-1] # La última fila y columna no son de interés
df.head()
```

```python
import matplotlib.pyplot as plt

# Suponiendo que df ya existe
languages = df.loc[:3, 'language']
locs = df.loc[:3, 'code']

# Función ajustada para barras horizontales
def add_labels(ax, y, x):
    for i in range(len(x)):
        ax.text(x[i] + max(x)*0.01, i, str(x[i]), va='center')  # Etiqueta al final de la barra

fig, ax = plt.subplots(1, 1, figsize=(12, 4))
bars = ax.barh(languages, locs, align='center')
ax.set_xlabel("Líneas de código")
ax.set_ylabel("Lenguaje")
ax.grid(which="both")
ax.set_title("Líneas de código (LdC) por lenguaje")

# Llamada para agregar etiquetas
add_labels(ax, languages, locs)

plt.show()
```
